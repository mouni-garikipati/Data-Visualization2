# **HANDLING MISSING VALUES**

```python
df.isnull().sum().sum()
```

**There are total 5157 missing values we need to replace or remove the values to make data more accurate for future data visualization now let us find in which column having the null values**

```python
df.isnull().sum()
```

**In above dataset ADDRESSLINE2 consists of 2521 null values, STATE consists of 1486 null values, TERRITORY consists on 1074 null values, POSTALCODE consists of 76 null values we need to clean that null values**    

```python
df["TERRITORY"]=df["TERRITORY"].fillna(df["TERRITORY"].mode()[0])
df["POSTALCODE"]=df["POSTALCODE"].fillna(df["POSTALCODE"].mode()[0])
```

```python
df.drop("ADDRESSLINE2", axis=1,inplace=True)
df.drop("STATE", axis=1,inplace=True)
```

```python
df
```

**Frome above dataset we have removed and replaced the null values ADDRESSLINE2, STATE the both the columns are remove because they havimg more than 50% of values are null values so they are not much use full for the future data visualization. TERRITORY and POSTALCODE are replaced with mode values because they are categorical data. where as categorical data is replaced with mode value and numerical data is replaced with mean and median. where as we are not using mean and median because we don't have missing values in numerical datatype**
